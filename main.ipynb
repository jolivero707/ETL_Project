```python
import json
import pandas as pd
from pprint import pprint
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()

# with open('C:/Users/javie/Documents/RICE/ETL_Project/SP.DYN.LE002.json') as json_data:
with open('SP.DYN.LE002.json') as json_data:

    d = json.load(json_data)
    
    pprint(d[0])
    print(d[0]['country']['value'])  
```

```python
country_list = []
country_codes = []
years = []
values = []

for row in d:
    try:
        country = row['country']['value']
        country_list.append(country)
        country_code = row["countryiso3code"]
        country_codes.append(country_code)
        year = row["date"]
        years.append(int(year))
        value = row["value"]
        values.append(value)

    except:
        continue

p_df = pd.DataFrame({"Country Name": country_list, "country_code": country_codes, "Year": years, "life expentancy": values})
pd.to_numeric(p_df["Year"])
p_df.head(100)
```
